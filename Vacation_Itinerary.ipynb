```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key

# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Store the CSV you saved created in part two into a DataFrame
vacation_data_df = pd.read_csv("weather_data/WeatherPy_vacation.csv")
vacation_data_df
```

```python
vacation_data_df = vacation_data_df.dropna()
vacation_data_df
```

```python
# Filter the DataFrame for each city you want to go to and create separate DataFrames for each city.

desired_cities = ["Salinopolis", "Cururupu", "Tutoia", "Itarema", "Aquiraz"]


travel_cities_df = vacation_data_df[vacation_data_df["City"].isin(desired_cities)]

travel_cities_df
```

```python
# Prepare Latitude-longitude pairs by iterating through the DataFrame.
coordinates = []
lat_lng = zip(travel_cities_df["Lat"], travel_cities_df["Lng"])

coordinates = list(lat_lng)
coordinates
```

```python
fig = gmaps.figure(center=(-3, -40), zoom_level=5)
```

```python
#Create template for markers
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
```

```python
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in travel_cities_df.iterrows()]


# Add a direction layer the desired vacation spots and a pop-up marker for each city.
locations = travel_cities_df[["Lat", "Lng"]]

marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

fig.add_layer(marker_layer)

fig
```

```python
# Add a direction layer the desired vacation spots and a pop-up marker for each city.

dir_layer = gmaps.directions_layer(coordinates[0], coordinates[1], 
                                   waypoints = [coordinates[2], coordinates[3], coordinates[4]],
                                   travel_mode='DRIVING')

fig.add_layer(dir_layer)
```
